```python
from spyplotter import Spectrum
import numpy as np
import matplotlib.pyplot as plt
```

# Read simulated spectrum

```python
spectrum_norm = Spectrum.from_powr(filepath='test_data/T33_logg4.0_Mdot-8.5_logL4.72_v1994_vmic8_Z0.1/formal.plot',keywords=['OPT'],dataset=1)
```

# Simulate noisy spectrum

```python
noisy_spectrum = Spectrum(spectrum_norm.x,spectrum_norm.y + np.random.normal(0, 0.005, len(spectrum_norm.x)))
print(noisy_spectrum.x_unit)
noisy_spectrum.plot();
plt.tight_layout()
```

# Export to a csv file

```python
import csv
# Save the noisy spectrum to a CSV file
output_file = "test_data/simulated_spectrum.csv"

with open(output_file, "w", newline="") as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(["Wavelength ", "Normalized Flux"])

    for l, amp in zip(noisy_spectrum.x.value, noisy_spectrum.y.value):
        csvwriter.writerow([l, amp])

print(f"Noisy spectrum has been saved to {output_file}.")
```
