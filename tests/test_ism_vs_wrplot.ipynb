```python
from spyplotter import ISMModel
from spyplotter import Spectrum
import astropy.units as u

from spyplotter.ism_lines import HLymanA
%matplotlib widget
```

# Test Lorentz profile

```python
ism_data = {
    "FeII": {
        "gas": {
            "n": 1e15,                 # cm^-2
            "vrad":-45
#             "T": 80000,                 # K
#             "vturb": 5.0,              # km/s
#             "A": 55.845                # atomic mass
        },
        "lines": {
            1608: {
                "lambda": 1608.451,
                "f": 5.8e-1,
                "gamma": 320000000
            },
        }
    }
}
```

```python
ism = ISMModel(ism_data)
li = ism.to_LineIdentifier(vrad=-45*u.km/u.s)
sp = Spectrum.from_powr(
    filepath='../tests/test_data/T33_logg4.0_Mdot-8.5_logL4.72_v1994_vmic8_Z0.1/formal.plot',
    keywords=['UV'],
    dataset=1
)
```

```python
#Read from wrplot created file
wrplot = Spectrum.from_file('../tests/test_data/ism_lorentz.dat')
```

```python
ax = wrplot.plot(label='wrplot')
ax.set_xlim(1607,1610)
li.plot(ax=ax)
ism.plot(ax=ax,sp=sp,label='spyplotter',lw=5,alpha=0.3)
sp.plot(ax=ax,label='without ISM')
ax.legend()
```

# Voigt Profile

```python
ism_data = {
    "FeII": {
        "gas": {
            "n": 1e13,                 # cm^-2
            "vrad":60,
            "T": 8000,                 # K
            "vturb": 10,              # km/s
            "A": 55.845                # atomic mass
        },
        "lines": {
            1608: {
                "lambda": 1608.451,
                "f": 0.7 ,
                "gamma": 500000000
            },
        }
    }
}
```

```python
ism = ISMModel(ism_data)
li = ism.to_LineIdentifier(vrad=60*u.km/u.s)
sp = Spectrum.from_powr(
    filepath='../tests/test_data/T33_logg4.0_Mdot-8.5_logL4.72_v1994_vmic8_Z0.1/formal.plot',
    keywords=['UV'],
    dataset=1
)
```

```python
sp.x_lim
```

```python
#Read from wrplot created file
wrplot = Spectrum.from_file('../tests/test_data/ism_voigt.dat')
```

```python
ax = wrplot.plot(label='wrplot')
ax.set_xlim(1607,1610)
li.plot(ax=ax)
ism.plot(ax=ax,sp=sp,label='spyplotter',lw=5,alpha=0.3)
sp.plot(ax=ax,label='without ISM')
ax.legend()
```
