# Plot population numbers

When just giving a string with an element, all population numbers for all ionisation stages are plotted. One can also give some more kwargs (like here the color) that are used in all plotting functions for the population numbers. The text is then also printed in the same color. This can be helpful when a specific set of population numbers is compared with the ones from another model.

```python
m1.popnums['HI'].shape
```

```python
m1.plot_popnums(element='C',use_marker=True,color="C00")
```

It is also possible to give the function a string with only one ionization state or a list of ionisation stages

By default, the ionisation stage corresponding to the population number is put as a text in the plot (just as in wrplot) but it is also possible to put the text in a legend by changing the text_in_legend parameter to True
If use_marker is set to False, a line instead of markers are used 

```python
m1.plot_popnums(element=['CII','CIII','CIV'],use_marker=False,text_in_legend=True)
plt.tick_params(direction="in")
```

## SED

To learn more about what can be done with SEDs (as comparing to observations) and reddening, read the docs in SED.ipynb

### SED with vs without lines

By default, the SED with lines is read in from a PoWR model but it can also be compared to the one without lines 

```python
# Read an SED from the model by also giving the keyword and specifying if the one with or without lines should be used (by default with lines)
# By default, the SED is scaled to 10 pc.
sed_10pc =m1.sed(keywords='IUE SHORT',with_lines=False)
ax=sed_10pc.plot(label='SED@10pc without lines')

# The SED can also be scaled to another distance by explicitly giving a distance (Unit is mandatory)
sed_55kpc =m1.sed(keywords='IUE SHORT')
ax=sed_55kpc.plot(ax=ax,label='SED@10kpc with lines')

# Set y axis to log scale
ax.set_yscale('log')
ax.legend()
```

### Scale SEDs to right distance

```python
# Read an SED from the model by also giving the keyword and specifying if the one with or without lines should be used (by default with lines)
# By default, the SED is scaled to 10 pc.
sed_10pc =m1.sed(keywords='IUE SHORT',with_lines=False)
ax=sed_10pc.plot(label='SED@10pc without lines')

# The SED can also be scaled to another distance by explicitly giving a distance (Unit is mandatory)
sed_55kpc =m1.sed(keywords='IUE SHORT',scale2distance=55*u.kpc)
ax=sed_55kpc.plot(ax=ax,label='SED@55kpc with lines')

# Set y axis to log scale
ax.set_yscale('log')
ax.legend()
```
