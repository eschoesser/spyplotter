```python
from spyplotter import Spectrum
import numpy as np
import matplotlib.pyplot as plt
from astropy import units as u
```

# Initialization of a Spectrum

You can either create a spectrum from two arrays directly:

```python
spectrum = Spectrum([1,2,3],[5,6,7])
```

```python
spectrum = Spectrum.from_powr(filepath='tests/test_data/T33_logg4.0_Mdot-8.5_logL4.72_v1994_vmic8_Z0.1',keywords=['OPT'],dataset=3)
```

# Plot spectrum

```python
fig = spectrum.plot(x_unit=u.eV);
ax = fig.get_axes()
```
